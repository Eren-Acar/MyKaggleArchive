```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
df.head(3)
```

```python
# Histogram for CGPA
plt.figure(figsize=(10, 6))
sns.histplot(df['cgpa'], kde=True, bins=20)
plt.title('Distribution of CGPA')
plt.xlabel('CGPA')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Histogram for Placement Exam Marks
plt.figure(figsize=(10, 6))
sns.histplot(df['placement_exam_marks'], kde=True, bins=20)
plt.title('Distribution of Placement Exam Marks')
plt.xlabel('Placement Exam Marks')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Histograms for CGPA and Placement Exam Marks by Placement Status
fig, axes = plt.subplots(2, 2, figsize=(14, 12), sharey=True)

# CGPA for placed=1
sns.histplot(df[df['placed'] == 1]['cgpa'], kde=True, bins=20, ax=axes[0, 0], color='green')
axes[0, 0].set_title('CGPA Distribution for Placed')
axes[0, 0].set_xlabel('CGPA')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].grid(True)

# CGPA for placed=0
sns.histplot(df[df['placed'] == 0]['cgpa'], kde=True, bins=20, ax=axes[0, 1], color='red')
axes[0, 1].set_title('CGPA Distribution for Not Placed')
axes[0, 1].set_xlabel('CGPA')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].grid(True)

# Placement Exam Marks for placed=1
sns.histplot(df[df['placed'] == 1]['placement_exam_marks'], kde=True, bins=20, ax=axes[1, 0], color='green')
axes[1, 0].set_title('Placement Exam Marks Distribution for Placed')
axes[1, 0].set_xlabel('Placement Exam Marks')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].grid(True)

# Placement Exam Marks for placed=0
sns.histplot(df[df['placed'] == 0]['placement_exam_marks'], kde=True, bins=20, ax=axes[1, 1], color='red')
axes[1, 1].set_title('Placement Exam Marks Distribution for Not Placed')
axes[1, 1].set_xlabel('Placement Exam Marks')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Scatter plot for CGPA vs Placement Exam Marks with 'placed' category
plt.figure(figsize=(10, 6))
sns.scatterplot(x='cgpa', y='placement_exam_marks', hue='placed', data=df, palette='Set2')
plt.title('CGPA vs Placement Exam Marks by Placement Status')
plt.xlabel('CGPA')
plt.ylabel('Placement Exam Marks')
plt.legend(title='Placed', loc='upper left')
plt.grid(True)
plt.show()
```
